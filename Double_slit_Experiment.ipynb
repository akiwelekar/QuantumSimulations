```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Parameters
num_particles = 2000
screen_width = 800
slit_distance = 100  # Distance between slits (pixels)
wavelength = 50      # Wavelength of the wave (arbitrary units)
screen = np.linspace(-screen_width // 2, screen_width // 2, screen_width)

# Wave function for each slit (Gaussian modulated wave)
def wave(x, slit_center):
    return np.exp(-(x - slit_center)**2 / 2000) * np.exp(1j * 2 * np.pi * (x - slit_center) / wavelength)

# Total wavefunction from two slits
psi = wave(screen, -slit_distance // 2) + wave(screen, slit_distance // 2)

# Probability distribution
probability = np.abs(psi)**2
probability /= probability.sum()  # Normalize

# Generate particle hits according to the probability distribution
particle_positions = np.random.choice(screen, size=num_particles, p=probability)

# Create animation frames
fig, ax = plt.subplots(figsize=(10, 4))
ax.set_xlim(-screen_width // 2, screen_width // 2)
ax.set_ylim(0, 100)
ax.set_title("Double Slit Experiment Simulation")
scat = ax.scatter([], [], s=1)

positions = []

def update(frame):
    positions.append(particle_positions[frame])
    ax.clear()
    ax.set_xlim(-screen_width // 2, screen_width // 2)
    ax.set_ylim(0, 100)
    ax.set_title("Double Slit Experiment Simulation")
    ax.scatter(positions, np.random.uniform(0, 100, size=len(positions)), s=1, color='blue')

ani = animation.FuncAnimation(fig, update, frames=num_particles, interval=1, repeat=False)

# Save animation
gif_path = "/content/double_slit_simulated_detection.gif"
ani.save(gif_path, writer='pillow', fps=60)

gif_path
```
